## Equilibrium analysis

In the article "DC motor First order approximation" it has benn showed how to characterize a DC motor using basic parameters provided in the motor Datasheet and predict the speed and responsiveness of motor for any given input voltage.

Let's look at some of parameters in mode detail under steady state conditions with a load. By applying a voltage and wait till the motor achieves constant velocity.

If a voltage source is applied to the motor terminal and a mechanical load torque $\tau$ on the rotor

$$
b \dot{\theta}=K_{t} i-\tau
$$

$$
R i=V-K_{b} \dot{\theta}
$$

Solving it is get

$$
i=\frac{b \dot{\theta}+\tau}{K_{t}}
$$

$$
V=R \frac{b \dot{\theta}+\tau}{K_{t}}+K_{b} \dot{\theta}
$$

$$
\dot{\theta}\left(\frac{R b}{K_{t}}+K_{b}\right)=V-\frac{R_{* \tau}}{K_{t}}
$$

$$
\dot{\theta}\left(b+\frac{K_{b} K_{t}}{R}\right)=\frac{V K_{t}}{R}-\tau
$$

Following

$$
\tau=\frac{V K_{t}}{R}-\dot{\theta}\left(b+\frac{K_{b} K_{t}}{R}\right)
$$

$$
\dot{\theta}=\left(\frac{V K_{t}}{R}-\tau\right)\left(b+\frac{K_{b} K_{t}}{R}\right)^{-1}
$$


* Slope of torque speed graph indicates inverse proportionality due to negative sign.
* The linear relationship between the torque provided by a motor and the speed at which it operates is specific to the input voltage at the motor's leads.

Thus it can be ploted the load vs speed graph at specified voltage and power specifications.
Lets plot the torque $\mathrm{v} / \mathrm{s}$ speed graph for nominal voltage of $12 \mathrm{~V}$ and let load torque vary from $0$ to $50$ $\mathrm{mNm}$ and observe the RPM and power requirement of the motor.

* At zero torque, the speed is highest. This speed is called no-load speed $\dot{\theta}_{n}$. The no-load speed can easily be calculated from the applied voltage and the speed constant of the motor.
* Enhancing the load torque leads to a linear reduction of the speed.
* Increasing the torque further reduces speed up to the point where the motor stops. The corresponding torque is called stall torque $\tau_{s}$
* The current corresponding to the stall torque is named starting current or no load current $i_{o}$
* A higher voltage results in a higher no-load speed
* The speed-torque gradient is unaffected with increase in voltage. Changing the applied voltage results in a parallel shift of the speed-torque line
* A strong motor is characterized by a flat speed-torque line. The speed drops only slightly when the load torque is enhanced. The value of the gradient is small.
* On a weaker motor the speed drop is larger. The speed-torque line is steeper and the value of its gradient is high.


## Stall Torque

The torque provided by the motor when the rotor no longer turns is known as the stall torque. The torque of motor at zero speed. It can be calculated as

$$\tau_{s}=\frac{V K_{t}}{R}$$

## No Load Speed

The speed at which the rotor turns when no torque is provided is known as the no-load speed. The no-load speed is called as such since it is the speed at which the rotor will turn when there is no load connected.

$$
\dot{\theta}=\left(\frac{V * K_{t}}{R}\right)\left(b+\frac{K_{b} K_{t}}{R}\right)^{-1}
$$

No load speed is maximum speed that can be attained by the motor at the specified voltage

## Stall Current

The stall current is the current drawn by the motor when rotor no longer turns. The current draw is proportional to the torque. The stall current is the maximum current drawn by the motor under equilibrium condition.

$$
i=\frac{\tau}{K_{t}} i=\frac{V}{R}
$$

## No Load Current

$$i=\frac{b \hat{\theta}}{K_{t}}$$

## Equilibrium Torque, Speed & Current

For a fixed voltage the equilibrium torque,speed and current can be expressed as

$$
\tau=\tau_{s}-\dot{\theta}\left(b+\frac{K_{b} K_{i}}{R}\right) \dot{\theta}=\theta_{n}-\tau\left(b+\frac{K_{b} K_{t}}{R}\right)^{-1} I=I_{S}-\dot{\theta}\left(\frac{K_{b}}{R}\right)
$$

## Motor Datasheet

Not all motor manufactures provide parameters that can be used directly to model the motor.
The parameter like $L$,$R$, torque constant, back-emf constant etc are called as Intrinsic Parameters of the motor.

The values like stall torque, stall current,no load speed and current are the values measured values and most often found in datasheets.

For example the important specifications of motor are often given in the form

* Nominal voltage $4.5 \mathrm{~V}$
* No load speed 23000 rpm
* No load current $70 \mathrm{~mA}$
* Stall torque of $0.34 \mathrm{mNm}$

Let's assume that it is provided also some intrinsic parameters like $\mathrm{R}$


It is known that the stall torque $\tau_{s}=\frac{V K_{t}}{R}$

Thusit could be computed $K_{t}$ from the equation above. 

The no load speed is given by

$$
\begin{aligned}
\dot{\theta_{n}}=&\left(\frac{V K_{t}}{R}\right)\left(b+\frac{K_{b} K_{t}}{R}\right)^{-1} \\
& \frac{\tau_{s}}{\dot{\theta}_{n}}=\left(b+\frac{K_{b} K_{t}}{R}\right)
\end{aligned}
$$

Viscous friction coefficient $b$ can be computed from this equation.

Just by observing the graph it could be determined the slope and the intercept ie $\left(b+\frac{K_{b} K_{t}}{R}\right)$ and $\frac{V K_{t}}{R}$.

Thus knowing the nominal voltage and estimate of $\mathrm{R}$, $K_{t}$ and $b$ could be determined from the graph itself. Thus all the information needed for the first order model of a DC Motor is available.

For example for the portscape [216 E motor datasheet](https://www.marutsu.co.jp/contents/shop/marutsu/ds/26n58_specifications.pdf) it is found that simple armature resistance yeilded a value of $15.2 ~\Omega$ and based on that given specifications and above formulat the intrinsic model parameters computed are

```python
Kt =0.0236
b =9.9544e-07
```

These are approximately equal to the values in the datasheet.


## Mechanical Power

The mechanical power $P$ delivered by the motor is $P=\tau \dot{\theta} P=\tau_{s} \dot{\theta}-\dot{\theta}^{2}\left(b+\frac{K_{b} K_{t}}{R}\right)$

The velocity that provides the maximum power could be finded by computing the gradient wrt $\dot{\theta}$

$$
\theta_{\max }=0.5 \tau_{s}\left(b+\frac{K_{b} K_{t}}{R}\right)^{-1}
$$

$$
\theta_{\max }=0.5 \dot{\theta_{n}}
$$

Thus maximum mechanical power is achieved when motor is running at half the no load velocity.

At this point the torque is $\tau_{\max }=0.5 \tau_{s}$

Thus maximum power is achieved when the torque is half of stall torque.

Since the current draw is proportional to the torque, then the current at the point of maximum power transfer shall also be one half of the stall current.

$$I_{\max }=0.5 I_{s}$$

The maximum mechanical power produced by the motor is as follows $P_{\max }=\frac{1}{4} \tau_{s} \dot{\theta}_{n}$

Thus maximum efficiency is obtained when speed is half no load speed and torque is half stall torque.

Power delivered by the motor is area of a rectangle under the torque/speed curve with one corner at the origin and another corner at a point on the curve.

It could be seen, the __maximum area__ is obtained when

$$
\theta_{\max }=0.5 \dot{\theta_{n}}
$$

and

$$\tau_{\max }=0.5 \tau_{s}$$

corresponding to the peak in the power vs speed curve

## Motor Efficiency

The efficiency $\eta$ of a motor is defined as the ratio between the input electrical power and output mechanical power

$$
\boldsymbol{\eta}=\frac{\boldsymbol{p}}{V I}
$$

$$
I=\frac{V_{s}}{R}-\dot{\theta}\left(\frac{K_{b}}{R}\right) P_{i}=V I=V \frac{V_{S}}{R}-V \dot{\theta}\left(\frac{K_{b}}{R}\right) \tau=\tau_{s}-\dot{\theta}\left(b+\frac{K_{b} K_{t}}{R}\right)
$$

$$
P_{o}=\frac{K_{t} V_{s}}{R} \dot{\theta}-\dot{\theta}^{2}\left(b+\frac{K_{b} K_{t}}{R}\right)
$$

$$
\dot{\theta}_{m e}=\frac{b c-\sqrt{b^{2} c^{2}-a b c d}}{b d}
$$

where:

* $\boldsymbol{a}=\tau_{s}-K t * i_{o}$
* $\boldsymbol{b}=\boldsymbol{b}+\frac{K_{b} K_{t}}{\boldsymbol{R}}$
* $c=\frac{V_{*} V_{s}}{R}$
* $\boldsymbol{d}=\boldsymbol{V}\left(\frac{K_{b}}{R}\right)$

## MATLAB code

The code in MATLAB to compute and plot the various DC motor curves

```python
%%% MATLAB code
%%% DC motor 2

clear;
model=1;
if model==1
    J = 6e-7; % moment of inertia kg.cm2
    b = 0     ; % FRICTIONAL DAMPING
    Kt = 23.90e-3 ; %torque constant Nm/A
    Kb = 23.90e-3;  %back emf constant mV/rpm
    R = 10    ;%rotor resistance ohms
    L = 0.8e-3; %rotor inductance

    Tf=0.38e-3; %frrictional torque
    io=16e-03; % no load current
    gain=12;
elseif model==0
    J = 6e-7; % moment of inertia kg.cm2
    b = 0     ; % FRICTIONAL DAMPING
    Kt = 12.30e-3 ; %torque constant Nm/A
    Kb = 12.30e-3;  %back emf constant mV/rpm
    R = 2.5    ;%rotor resistance ohms
    L = 0.22e-3; %rotor inductance

    Tf=0.38e-3; %frrictional torque
    io=31e-03; % no load current
    gain=6;
elseif model==2
    J = 6e-7; % moment of inertia kg.cm2
    b = 0     ; % FRICTIONAL DAMPING
    Kt = 25.80e-3 ; %torque constant Nm/A
    Kb = 25.80e-3;  %back emf constant mV/rpm
    R = 15.2    ;%rotor resistance ohms
    L = 1e-3; %rotor inductance

    Tf=0.38e-3; %frrictional torque
    io=15e-03; % no load current
    gain=15;
    elseif model==3
    J = 6e-7; % moment of inertia kg.cm2
    b = 0     ; % FRICTIONAL DAMPING
    Kt = 33.50e-3 ; %torque constant Nm/A
    Kb = 33.50e-3;  %back emf constant mV/rpm
    R = 32    ;%rotor resistance ohms
    L = 1.50e-3; %rotor inductance

    Tf=0.38e-3; %frrictional torque
    io=20e-03; % no load current
    gain=24;
end

%% Second order DC motor model
s = tf('s');
P_motor = Kt/((J*s+b)*(L*s+R)+Kt*Kb);
disp('\n\nDC motor Step response parameters');
S = stepinfo(P_motor)
%ltiview('step', P_motor, 0:0.1:5);
%[n,d]=tfdata(P_motor)
[a2,b1,c1]=zpkdata(P_motor);
dom=max(b1{1});
%p=poles(sys(P_motor));
disp(sprintf('Dominant Poles of DC motor  %0.03f',dom));

%% First order DC motor model
P_first = (Kt/R)/(J*s+b+(Kt*Kb/R));
%p=poles(sys(P_first));
%disp(sprintf('Poles of first DC motor model %0.03f  ',p(0)));
disp('\n\nFirst order model DC motor Step response parameters');
S = stepinfo(P_first)
pole1=(b+(Kt*Kb/R))/J;
disp(sprintf('Poles of first DC motor model %0.03f  ',pole1));

%ltiview('step', P_first, 0:0.1:5);

%% Voltage vs Speed
v=0:0.1:24;
angle=(v-R*io)*(Kt/(R))/((b+(Kt*Kb/R)));

figure(1);
xmarkers = [1,6,12,15,24]; % place markers at these x-values
ymarkers = (xmarkers-R*io)*(Kt/(R))/((b+(Kt*Kb/R)));


y1=ymarkers*60/(2*3.1412);
%plot(v,angle*60/(2*3.1412),'r')

plot(v,angle*60/(2*3.1412),'m',xmarkers,y1,'b*')
hold on
strValues = strtrim(cellstr(num2str([xmarkers(:) y1(:)],'(%d,%d)')));
text(xmarkers,y1,strValues,'VerticalAlignment','bottom');

grid on
xlabel('\bf voltage');
ylabel('\bf RPM');
title('\bf Voltage vs RPM');
hold off


%% torque speed characteristics
%Tf=0;
Tff=0;
Vs = gain-R*io;
t=0:0.05:30;
theta=((Vs*Kt/R)-(t/1000)-Tff)/(b+(Kb*Kt/R));
theta1=theta*60/(2*3.1412);
theta=theta1(theta1>0);
t=t(theta1>0);
figure(2);
power=theta.*((t/1000)+Tff)*(1000*2*3.1412/60);



%% Stall torque,current and no load speed computation

Tff=0;
stallTorque=(Vs*Kt/R)-Tff;
stallCurrent=(Vs/R);
OmegaNL = ((Vs*Kt/(R))-Tff)/( b + (Kt*Kt)/R);

fprintf('stall torque %0.3f\n',stallTorque);

fprintf('stall current %0.3f\n',stallCurrent);

fprintf('no load velocity %0.3f\n',OmegaNL*60/(2*3.1412));

hold on
Y=[0;stallTorque/2;stallTorque/2;0];
X=[0;0;OmegaNL/2;OmegaNL/2];

XX=X;%(index([1 1:end end]));
YY=Y;%[0 Y(index) 0];
      



Kt=(stallTorque*R/gain);
fprintf('computed value of Kt %0.3f\n',Kt);
b=(stallTorque/OmegaNL)-(Kt*Kt/R);
fprintf('computed value of b %0.3f\n',b);


%%%% motor efficiency computation

theta=theta*2*3.1412/60;
I = (Vs-(Kt*theta))/R;
J = (Kt*I) - (Kt*io)-Tff; 
Po = (J.*theta);
Pi = gain*I;
Eff = (Po./Pi)*100;
%figure(3);


[xx,yy]=max(Eff);
maxEff=xx;
mostEffVelocity=theta(yy);
disp(sprintf('Max Motor Efficiency : %f Percent ',100*maxEff));
disp(sprintf('Velocity at max efficiency %f rads/sec ',mostEffVelocity*60/(2*3.1412)));



%% plotting the motor characteristic curves
figure(2)
fill(XX*60/(2*3.1412),YY*1000,'b');
%


theta=theta*60/(2*3.1412);
[hAx,hLine1,hLine2] = plotyy(theta,t,theta,power);
Eff=Eff/2;
title ('\bf DC Motor Torque-Speed Characteristic'); 
xlabel('Speed')
ylabel(hAx(1),'Torque mNm') % left y-axis
ylabel(hAx(2),'Power mW') % right y-axis
h3=plot(theta,Eff,'r');
h4=plot(angle*60/(2*3.1412),v,'m',y1,xmarkers,'b*')

hold on
strValues = strtrim(cellstr(num2str([y1(:) xmarkers(:)],'(%d,%d)')));
text(y1,xmarkers,strValues,'VerticalAlignment','bottom');

legend([hLine1;hLine2;h3(1);h4(1)],'torque','power','efficiency','voltage');
%legend(h3(1),'efficiency')

%legend(h4(1),'voltage')

%axis([0 OmegaNL*60/(2*3.1412) 0 100]);
grid on;


%o = theta;
a11 = Kt*Vs/R-Kt*io;
b1 = ((Kt*Kt/R));
c1 = gain*Vs/R;
d1 = gain*Kt/R;
mostEffVelocity = ((b1*c1) - sqrt((b1*b1*c1*c1)-(a11*b1*c1*d1)))/(d1*b1);
disp(sprintf('Theorotical Velocity at max efficiency %f rads/sec ',mostEffVelocity*60/(2*3.1412)));
```
